#Recoding 

GENDERFACTOR: Will use 0 for Male, 1 for Female. 

ETHNICITYFACTOR: Will use 0 for Non-Hispanic, 1 for Hispanic.

RACEFACTOR: 0=white, 1=black, 2= asian or pacific islander, 3=other, 4=unknown

```r
allergy$GENDERFACTOR[allergy$GENDERFACTOR=='S0 - Male'] <- 0
allergy$GENDERFACTOR[allergy$GENDERFACTOR=='S1 - Female'] <- 1
```

```r
head(allergy)
```

```r
allergy$ETHNICITYFACTOR[allergy$ETHNICITYFACTOR=='E0 - Non-Hispanic'] <- 0
allergy$ETHNICITYFACTOR[allergy$ETHNICITYFACTOR=='E1 - Hispanic'] <- 1
```

```r
head(allergy)
```

```r
allergy$RACEFACTOR[allergy$RACEFACTOR=='R0 - White'] <- 0
allergy$RACEFACTOR[allergy$RACEFACTOR=='R1 - Black'] <- 1
allergy$RACEFACTOR[allergy$RACEFACTOR=='R2 - Asian or Pacific Islander'] <- 2
allergy$RACEFACTOR[allergy$RACEFACTOR=='R3 - Other'] <- 3
allergy$RACEFACTOR[allergy$RACEFACTOR=='R4 - Unknown'] <- 4
```
